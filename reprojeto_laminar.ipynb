```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
Q = [2, 4, 8, 16, 30] #L/min
QN = [i/100 for i in Q]
T = [879, 1180, 1812, 2609, 3499] #volts
dP = [3500*(((i/1000)-0.25)/3.75)**2 for i in T]
QMS = [i/60000 for i in QN]
a = 0.001**2
V = [i/a for i in QMS]

Re = [i*0.001/(15*math.pow(10,-6)) for i in V]

ECV = [1.2*0.5*(i**2) for i in V]
#energia cinética por unidade de volume

dPdx = [-i/0.08 for i in dP]

fzero = [-0.001/i for i in ECV]

f = [i*j for i,j in zip(fzero, dPdx)]
print(f)
print(Re)
plt.figure(figsize=(10,10))
plt.plot(Q, dP, 'o', color = "red")
plt.xlim(0,45)
plt.title("Pressão por vazão")
plt.grid()
```

```python
plt.figure(figsize=(10,10))
plt.plot(Re, f, '-')
plt.title("f por Re")
plt.grid()
```
